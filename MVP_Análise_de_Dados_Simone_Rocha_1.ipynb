###**4.1 Scatter Plot:**

```python
# Plotando scatter plot com cor baseada no 'World %'
plt.figure(figsize=(10, 6))
plt.scatter(
    dataset['Growth Rate'],
    dataset['Area (km2)'],
    c=dataset['World %'],
    cmap='coolwarm',
    alpha=0.7
)

# Incluindo título e rótulos dos eixos
plt.title('Taxa de Crescimento da População x Área (km²)', fontsize=14)
plt.xlabel('Taxa de Crescimento (%)', fontsize=12)
plt.ylabel('Área (km²)', fontsize=12)

# Aplicando escala logarítmica ao eixo Y (Área)
plt.yscale('log')

# Adicionando barra de cores
plt.colorbar(label='Porcentagem da População Mundial (%)')

# Exibindo o gráfico
plt.show()
```

O gráfico exibe a Taxa de Crescimento Populacional no eixo X e a Área dos países no eixo Y, utilizando uma escala logarítmica. A cor dos pontos representa a Porcentagem da População Mundial (%) de cada país, seguindo a paleta de cores "coolwarm".

Observa-se que a maioria dos pontos está concentrada na parte superior do gráfico, o que sugere que muitos países possuem áreas expressivas, acima de 100 km² (na escala logarítmica).

Para que o Scatter Plot seja exibido corretamente, é fundamental que cada coluna esteja no formato adequado de dados. Por exemplo, a coluna "Área (km²)", que contém valores como "M" (milhões) e "K" (milhares), está classificada como um tipo de objeto, mas deveria ser numérica. Da mesma forma, as colunas "Taxa de Crescimento" e "Porcentagem Mundial" precisam ser numéricas para possibilitar operações matemáticas e estatísticas precisas.

Portanto, uma nova análise será necessária após o pré-processamento dos dados, a fim de melhorar a qualidade dos resultados apresentados pelo Scatter Plot.

###**4.2 Gráfico de Linhas com Seaborn**

```python
# Preparando o dataset para o gráfico de linhas
# Pegaremos apenas as colunas de países e suas populações nos dois anos para traçar as linhas
data = dataset[['Country', 'Population 2024', 'Population 2023']].head(10)  # Filtrando os 10 primeiros países

# Colocando os dados no formato "long" para facilitar o uso no Seaborn
data_melted = data.melt('Country', var_name='Ano', value_name='População')

# Criando o gráfico de linhas
plt.figure(figsize=(10,6))
sns.lineplot(data=data_melted, x='Country', y='População', hue='Ano', marker="o")

plt.title('População dos 10 Países Mais Populosos (2023 vs 2024)')
plt.xticks(rotation=45)
plt.show()
```

###**5.1 Converter valores com "M" e "K"**

```python
# Função para converter valores com "M" e "K"
def convert_area(area_str):
    if isinstance(area_str, str):
        # Verifica se a string contém 'M' (milhões)
        if 'M' in area_str:
            return float(area_str.replace('M', '')) * 1e6
        # Verifica se a string contém 'K' (milhares)
        elif 'K' in area_str:
            return float(area_str.replace('K', '')) * 1e3
        # Verifica se a string contém '<' (menor que)
        elif '<' in area_str:  # Adiciona tratamento para valores menores que 1
            return 0.0  # Define um valor padrão para '< 1'
    # Retorna o valor original se não for uma string com 'M' ou 'K'
    return float(area_str)

# Aplicando a função na coluna 'Area (km2)'
dataset['Area (km2)'] = dataset['Area (km2)'].apply(convert_area)

# Verificando se a conversão foi bem-sucedida
print(dataset['Area (km2)'].head())
```

Aplica-se a função convert_area, para converter valores em uma coluna que contêm abreviações como "M" (milhões) e "K" (milhares) em seus valores numéricos correspondentes.

A função começa verificando se o valor da área é uma string. Se for, a função prossegue com as conversões; caso contrário, ela apenas retorna o valor original convertido para float.
Na conversão de valores com 'M' (milhões), considera-se que a string contém o caractere 'M', indicando que o valor está em milhões.

O código remove o 'M' da string usando o método replace, converte o restante do número para float, e multiplica o valor por 1 milhão (1e6) para transformar o valor na escala correta.

Sobre a string 'K', deduzimos que o valor está em milhares. O código remove o 'K' da string, converte o restante do número para float, e multiplica o valor por 1 mil (1e3).

Sobre a string que contém o caractere '<', que indica um valor "menor que", a função define o valor como 0.0 para simplificar o cálculo. Isso é útil para tratar valores como "<1". Assim, garantimos que os valores da coluna "Area (km2)" sejam convertidos para números reais, independentemente de estarem em milhões (M) ou milhares (K). Ao fazer isso,  padroniza-se os dados para que todos os valores estejam em uma unidade numérica, facilitando a análise e visualização de dados sem ambiguidade nos valores.

```python
# Definindo as variáveis que serão padronizadas
scaler = StandardScaler()

# Aplicando padronização nas colunas 'Growth Rate', 'Density (/km²)' e 'World %'
dataset[['Growth Rate', 'Density (/km2)', 'World %']] = scaler.fit_transform(
    dataset[['Growth Rate', 'Density (/km2)', 'World %']]
)

# Plotando o scatter plot com as variáveis padronizadas
plt.figure(figsize=(10, 6))
plt.scatter(
    dataset['Growth Rate'],  # 'Growth Rate' já está padronizado
    dataset['Area (km2)'],  # A 'Área (km2)' permanece não padronizada
    c=dataset['World %'],  # 'World %' padronizado para colorir os pontos
    cmap='coolwarm',
    alpha=0.7
)

# Incluindo título e rótulos dos eixos
plt.title('Taxa de Crescimento da População x Área (km²)', fontsize=14)
plt.xlabel('Taxa de Crescimento (padronizada)', fontsize=12)
plt.ylabel('Área (km²)', fontsize=12)

# Aplicando escala logarítmica ao eixo Y (Área)
plt.yscale('log')

# Adicionando barra de cores para o 'World %'
plt.colorbar(label='Porcentagem da População Mundial (padronizada)')

# Exibindo o gráfico
plt.show()
```

O gráfico de dispersão explora a relação entre a taxa de crescimento populacional, a área territorial dos países e sua porcentagem na população mundial. No eixo X, temos a taxa de crescimento padronizada, que facilita a comparação entre países com crescimento acima ou abaixo da média global. O eixo Y utiliza uma escala logarítmica para representar a área dos países, o que equilibra a visualização de nações grandes e pequenas. As cores indicam a porcentagem da população mundial, com tons mais quentes (vermelho/laranja) representando países com uma participação demográfica maior.

A escolha pela padronização das variáveis "Growth Rate", "Density (/km²)" e "World %" foi fundamental para permitir uma comparação justa entre os países, já que essas variáveis possuem escalas muito diferentes. A padronização transforma essas medidas em valores com média 0 e desvio padrão 1, de modo que cada país pode ser analisado em relação à média global de crescimento populacional e densidade. Isso significa que no eixo X, um país com um valor positivo apresenta uma taxa de crescimento populacional acima da média global, enquanto valores negativos indicam um crescimento populacional inferior à média.

No gráfico, não há uma correlação direta entre o tamanho geográfico e o crescimento populacional, sugerindo que outros fatores influenciam o crescimento. Países pequenos como Singapura podem ter altas taxas de crescimento, enquanto países grandes como Rússia podem crescer lentamente. Além disso, outliers — países com alta densidade populacional e rápido crescimento — são visíveis, destacando nações com desafios específicos de infraestrutura e recursos.

Assim, o gráfico oferece uma visão rica sobre as dinâmicas de crescimento populacional e como o crescimento, a densidade e a área interagem em diferentes contextos globais, revelando padrões e discrepâncias importantes na distribuição populacional mundial.
